```python
def average(c):
    total = 0
    for i in c:
        try:
            total+=i
        except:
            print(i)
    return total/len(c)

def correlation(c1,c2):
    avg1 = average(c1)
    avg2 = average(c2)
    
    numerator = 0
    for i in range(len(c1)):
        numerator+=(c1[i]-avg1)*(c2[i]-avg2)
        
    d1= 0
    for i in range(len(c1)):
        d1 += (c1[i]-avg1)**2
    d2= 0
    for i in range(len(c2)):
        d2 += (c2[i]-avg2)**2
    denominator = (d1*d2)**.5
    return numerator/denominator

l = list(df2)
unwanted = ['generation','gdp_per_capita ($)','HDI for year','country-year','country','suicides/100k pop','suicides_no']
wanted = ['population','sex','age','year','suicides/100k pop']
for c1 in l:
    if c1 not in wanted:
        continue
    print(c1, end = " ")
print()
for c1 in l:
    if c1 not in wanted:
        continue
    #print(c1, end="")
    for c2 in l:
        if c2 not in wanted:
            continue
        print(f' {correlation(df2.loc[:,c1].to_numpy(),df2.loc[:,c2].to_numpy()): .9f}', end=" " )
    print()
```
